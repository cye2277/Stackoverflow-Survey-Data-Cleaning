**<big>Corresponding Question</big>**  
   
   
**YearsCoding: Including any education, for how many years have you been coding?**    
**YearsCodingProf: For how many years have you coded professionally (as a part of your work)?**  
**ConvertedSalary: Salary converted to annual USD salaries using the exchange rate on 2018-01-18, assuming 12 working months and 50 working weeks.**

**1. Check the number of na values in each column.**

```python
df.isnull().sum()
```

**2. Delete the records that contain na values in the two related columns.**

```python
year_info = df.dropna(subset=['YearsCoding', 'YearsCodingProf'])
```

**3. Check the differences made by the dropna command.**

```python
year_info.isnull().sum()
```

```python
year_info.shape
```

**4. Take a closer look at the Converted Salary column.**

```python
year_info['ConvertedSalary'].describe()
```

**5. Start out to clean the Converted Salary data.**

```python
year_info[year_info['ConvertedSalary'] < 100][['Respondent', 'ConvertedSalary']]
```

**We can see that there are 74 respondents use a number less than 100 as their converted annual salary, which is apparently unreasonable, so we decided to drop these records.**

```python
year_info = year_info[year_info['ConvertedSalary'] >= 100]
```

```python
year_info.shape
```

**The following two kinds of records don't look like someone's exact salary, but it is possible that some respondents just choose these round numbers as an approximate answer to this question.**

```python
year_info[year_info['ConvertedSalary']==100000.0].shape[0]
```

```python
year_info[year_info['ConvertedSalary']==1000000.0].shape[0]
```

**Since the maximum option of this question is 2 million, it is possible that some respondents choose this option deliberately even if their actual salary is less than this.**

```python
year_info[year_info['ConvertedSalary']==2000000].shape[0]
```

**As a result, we decide to drop the records who has "0-2 years" professional coding experience and 2 million Converted Salary.**

```python
year_info[(year_info['ConvertedSalary']==2000000) & (year_info['YearsCodingProf']=='0-2 years')][['YearsCoding', 'YearsCodingProf', 'ConvertedSalary']]
```

```python
year_info = year_info[(year_info['ConvertedSalary']!=2000000) | (year_info['YearsCodingProf']!='0-2 years')]
```

```python
year_info.shape
```

**6. check for rows that contains too much na values**

```python
year_info['na_count'] = year_info.shape[1] - year_info.count(axis=1)
year_info.head()
```

```python
# make sure there is no missing value
year_info['na_count'].describe()
```

**We decide to drop the rows that contains more than 12 na values.**

```python
year_info[year_info['na_count'] > 12].shape[0]
```

```python
year_info = year_info[year_info['na_count'] <= 12]
```

```python
year_info.shape
```

**7. Start out to clean the YearsCoding and YearsCodingProf data.**

```python
year_info['YearsCoding'].unique()
```

```python
# create a new column so we can compare and sort the Years of coding value more easily:
year_info.loc[df['YearsCoding'] == '0-2 years', 'Coding_level'] = 1
year_info.loc[df['YearsCoding'] == '3-5 years', 'Coding_level'] = 2
year_info.loc[df['YearsCoding'] == '6-8 years', 'Coding_level'] = 3
year_info.loc[df['YearsCoding'] == '9-11 years', 'Coding_level'] = 4
year_info.loc[df['YearsCoding'] == '12-14 years', 'Coding_level'] = 5
year_info.loc[df['YearsCoding'] == '15-17 years', 'Coding_level'] = 6
year_info.loc[df['YearsCoding'] == '18-20 years', 'Coding_level'] = 7
year_info.loc[df['YearsCoding'] == '21-23 years', 'Coding_level'] = 8
year_info.loc[df['YearsCoding'] == '24-26 years', 'Coding_level'] = 9
year_info.loc[df['YearsCoding'] == '27-29 years', 'Coding_level'] = 10
year_info.loc[df['YearsCoding'] == '30 or more years', 'Coding_level'] = 11
```

```python
year_info['Coding_level'].unique()
```

```python
year_info['YearsCodingProf'].unique()
```

```python
# create a new column so we can compare and sort the Years of professional coding value more easily:
year_info.loc[df['YearsCodingProf'] == '0-2 years', 'CodingProf_level'] = 1
year_info.loc[df['YearsCodingProf'] == '3-5 years', 'CodingProf_level'] = 2
year_info.loc[df['YearsCodingProf'] == '6-8 years', 'CodingProf_level'] = 3
year_info.loc[df['YearsCodingProf'] == '9-11 years', 'CodingProf_level'] = 4
year_info.loc[df['YearsCodingProf'] == '12-14 years', 'CodingProf_level'] = 5
year_info.loc[df['YearsCodingProf'] == '15-17 years', 'CodingProf_level'] = 6
year_info.loc[df['YearsCodingProf'] == '18-20 years', 'CodingProf_level'] = 7
year_info.loc[df['YearsCodingProf'] == '21-23 years', 'CodingProf_level'] = 8
year_info.loc[df['YearsCodingProf'] == '24-26 years', 'CodingProf_level'] = 9
year_info.loc[df['YearsCodingProf'] == '27-29 years', 'CodingProf_level'] = 10
year_info.loc[df['YearsCodingProf'] == '30 or more years', 'CodingProf_level'] = 11
```

```python
year_info['CodingProf_level'].unique()
```

**We can see from above questions that it is unreasonable for a repondent's YearsCoding value less than YearsCodingProf value, so we check for these mistaken records.**

```python
# Now we can compare these two values:
year_info[year_info['Coding_level'] < year_info['CodingProf_level']].shape[0]
```

**Give a look at these records and drop them.**

```python
year_info[year_info['Coding_level'] < year_info['CodingProf_level']][['YearsCoding', 'YearsCodingProf', 'Coding_level', 'CodingProf_level']].head(10)
```

```python
year_info[year_info['CodingProf_level'] - year_info['Coding_level'] > 1][['YearsCoding', 'YearsCodingProf', 'Coding_level', 'CodingProf_level']]
```

```python
year_info = year_info[year_info['Coding_level'] >= year_info['CodingProf_level']]
```

```python
year_info.shape
```

```python
year_info.to_csv('CodingYearData.csv', index=False)
```

**<big>Corresponding Question</big>**    
    
    
**CompanySize: Approximately how many people are employed by the company or organization you work for?**  
**TimeFullyProductive: Suppose a new developer with four years of experience, including direct experience working with your company's main technical stack, joined your team tomorrow.  All other things being equal, how long would you expect it to take before they were fully productive and contributing at a typical level to your main code base?**    
**ConvertedSalary: Salary converted to annual USD salaries using the exchange rate on 2018-01-18, assuming 12 working months and 50 working weeks.**

```python
df.shape
```

```python
env_info = df.dropna(subset=['CompanySize', 'TimeFullyProductive'])
env_info.shape
```

**We first take a look at the CompanySize and TimeFullyProductive data.**

```python
# make sure there is no na value left
env_info['CompanySize'].unique()
```

```python
env_info['TimeFullyProductive'].unique()
```

**Then we clean the ConvertedSalary data as above.**

```python
env_info = env_info[env_info['ConvertedSalary'] >= 100]
```

```python
env_info.shape
```

```python
env_info = env_info[(env_info['ConvertedSalary']!=2000000) | (env_info['YearsCodingProf']!='0-2 years')]
```

```python
env_info.shape
```

```python
env_info.groupby('CompanySize')['ConvertedSalary'].describe()
```

```python
size_pts = env_info.groupby('CompanySize')['ConvertedSalary'].mean()

plt.figure(figsize=(12,5), dpi= 100)
plt.plot(size_pts, linewidth = 0.8, marker = '.', label = "Salary", color='#E11B00')

plt.xlabel("Company Size")
plt.ylabel("Converted Salary")
plt.ylim(60000, 140000)
x_order = ['Fewer than 10 employees','10 to 19 employees', '20 to 99 employees', '100 to 499 employees', '500 to 999 employees', '1,000 to 4,999 employees', '5,000 to 9,999 employees', '10,000 or more employees']
plt.xticks(np.arange(0, 8, 1), fontsize=9, horizontalalignment='center', rotation = 45, alpha=0.7)
# df.set_index('Company Size').loc(xorder)
plt.yticks(np.arange(60000, 140000, 20000), fontsize=12, alpha=.7)
plt.title("Converted Salary - Company Size", fontsize=16)
plt.grid(axis='both', alpha=.3)
plt.fill_between(size_pts.index, size_pts, 0, color = "#CE1141", alpha=0.3)
plt.legend()
```
